```python
d = {
    'col1': [1,2,4,3,2,1,6,7,3,9],
    'col2': [5,8,9,10,12,8,6,2,8,7]
}
df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'], 'k', label = 'Col1')
ax.plot(df2['col2'], 'ko--', label = 'Col2')
ax.legend(loc='best')
```

```python
#ticks, labels, and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum(), 'y--')
```

```python
ax.set_xticks([0,250,500,750,1000])
```

```python
ax
```

```python
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['One','Two','Three','Four','Five'], rotation = 30, fontsize = 'small')
ax.set_title('My Matplotlib Plot with Ticks')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
fig
```

```python
props = {
    'title': 'My Matplotlib Plot with Ticks',
    'xlabel': 'Stages',
    'ylabel': 'Values'
}
ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(randn(100).cumsum(), 'ko-', label = 'one')
ax.plot(randn(100).cumsum(), 'g--', label = 'two')
ax.plot(randn(100).cumsum(), 'y.', label = 'three')

ax.legend(loc = 'best')
```

```python
#saving a figure in an image file
fig.savefig('fig1.png')
```
